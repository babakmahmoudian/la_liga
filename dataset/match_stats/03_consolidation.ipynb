```python
import pandas as pd
```

# Joining the separate dataframes

Loading the original dataframes

```python
df_matches = pd.read_csv("./initial_ds/match_stat_links.csv")
df_stats = pd.read_csv("./initial_ds/match_stats.csv")
```

The dataframe containing the matches and the index columns (date and team names):

```python
df_matches
```

The dataframe containing match stats:

```python
df_stats
```

Joining the two initial dataframes on their common column which is the URL of the match's stats (stats_link):

```python
joined_df = pd.merge(left=df_matches, right=df_stats, on='stats_link', how='inner')

joined_df
```

# Cleaning up

```python
joined_df.columns
```

```python
joined_df = joined_df.drop(['season', 'link', 'stats_link'], axis='columns')
```

# Saving the results into file

```python
joined_df.to_csv("./processed_ds/match_stats.csv", index=False)
```
